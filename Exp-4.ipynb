## DATASET

```python
import pandas as pd

data = {'Square Footage': [1200, 1500, 2000, 1800, 1400, 1300, 1600],
        'Bedrooms': [2, 3, 4, 2, 2, 3, 3],
        'Bathrooms': [1, 2, 2, 1, 1, 2, 2],
        'Rent Amount': [1500, 2000, 2500, 2300, 1800, 1900, 2200],
        'Garage': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No']}

df = pd.DataFrame(data)
df
```

#### .replace()

```python
EncodingHM = {
    'Yes' : 1,
    'No' : 2
}


df['Garage Encoded'] = df['Garage'].replace(EncodingHM)
df
```

#### .get_dummies()

```python
df = pd.get_dummies(df, columns=['Bedrooms'])
df
```

#### sklearn Encoding

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
df['Garage_Encoded_2'] = encoder.fit_transform(df[['Garage']])
df
```
