3. Perform various imputation techniques for the missing data. 

    a. Mean-median imputation

    b. Arbitrary value imputation

    c. Frequent category imputation

    d. Missing category imputation

## DATASET

```python
import pandas as pd
import numpy as np


sales_data = {'date': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05', '2021-01-06'],
              'region': ['East', 'West', 'North', 'South', 'East', 'West'],
              'sales': [100, 200, np.nan, 200, 500, np.nan],
              'expenses': [50, np.nan, 75, 80, np.nan, 90],
              'profit': [np.nan, 150, 100, np.nan, 400, 300],
              'is_holiday': [0, 0, 1, 0, np.nan, 1]}

sales_df = pd.DataFrame(sales_data)
sales_df
```

#### Mean-median Imputation

```python
mean = sales_df['expenses'].median() #Mean or Median whichever value fits best for the dataset
sales_df['expenses'] = sales_df['expenses'].fillna(value=mean)
sales_df
```

#### Arbitrary value Imputation

```python
sales_df['profit'] = sales_df['profit'].fillna(value=50)
sales_df
```

#### Frequent category Imputation

```python
#Frequent category is also the MODE of the Data

mode = sales_df['sales'].mode()[0]

sales_df['sales'] = sales_df['sales'].fillna(mode)
sales_df
```

#### Missing category Imputation

```python
sales_df['is_holiday'] = sales_df['is_holiday'].fillna('unknown')
sales_df
```
